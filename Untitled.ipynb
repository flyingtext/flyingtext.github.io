```python
import sqlite3
import hanja
import unicodedata
```

```python
conn = sqlite3.connect('v2/static/data_new.sqlite')
cursor = conn.cursor()
cursor2 = conn.cursor()
```

```python
# hanja.translate('大韓民國은 民主共和國이다.', 'substitution')
cursor.execute('select distinct 처방명 from prescription_structure;')
rows = cursor.fetchall()

for row in rows :
    converted = hanja.translate(unicodedata.normalize('NFD', row[0]), 'substitution')
    print(converted)
    cursor.execute('update prescription_structure set 처방명한글="' + converted + '" where 처방명="' + row[0] + '"')
    
conn.commit()
```
